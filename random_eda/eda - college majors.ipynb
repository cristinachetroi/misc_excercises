#### DataFrame.plot()
When we call this method, we specify the data we want plotted as well as the *type of plot* - *kind* parameter

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', figsize = (8, 8), title = 'Employed vs. Sample size')
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize = (14, 30))

for i in range(len(cols)):
    ax = fig.add_subplot(8,2,i+1)
    ax = recent_grads[cols[i]].plot(kind = 'hist', title=cols[i], edgecolor='k')
```
