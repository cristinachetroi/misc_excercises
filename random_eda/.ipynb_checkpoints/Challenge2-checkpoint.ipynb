#### Question: 
What week in the last year had the greatest number of commits?

#### Result
Weeks 4 and 7 have highest number of commits (8) in 2017

```python
#import libraries
import numpy as np
import pandas as pd
#import seaborn as sns
#import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
#import 2017 commits data into a df
dataset = pd.read_json('https://api.github.com/repos/d3/d3/commits?since=2017-01-01T00:00:00Z&until=2017-12-31T23:59:59Z&per_page=100')
# display all text in a cell without truncation
pd.set_option('display.max_colwidth', -1)
```

```python
print(dataset.info())
dataset.head(1)
```

```python
#look closer into Commit column
dataset.commit[:15]
```

```python
#extrapolate 'commit' column (contains the date info) in a new df and a flatter formatting

commit = pd.read_json( (dataset['commit']).to_json(), orient='index')
```

```python
commit.head(20)
```

```python
#extrapolate either author or committer column into a new df
author = pd.read_json( (commit['author']).to_json(), orient='index')
```

```python
print(author.info())
author.head()
```

```python
#create a new week col from datetime column
author['week'] = author['date'].dt.week
```

```python
author.head(10)
```

```python
#count frequency distribution for each week
author.week.value_counts()
```

### Result: weeks 4 and 7 have highest number of commits (8)
